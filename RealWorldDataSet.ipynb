## store the data in a table

```python
url = r"https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv"
df = pd.read_csv(url)
df.to_sql("CHICAGO_SCHOOL_DATA",con, if_exists = 'replace', index = False, chunksize = 100)
```

### Query the database system catalog to retrieve table metadata

```python
%sql SELECT name FROM sqlite_master WHERE type='table' ;
```

```python
%sql select name, type, Length(type) from PRAGMA_TABLE_INFO ('CHICAGO_SCHOOL_DATA');
```

```python
%%sql select count(*) 
FROM CHICAGO_SCHOOL_DATA 
WHERE "Elementary, Middle, or High School" = 'ES' ;
```

```python
%sql select MAX(SAFETY_SCORE) from CHICAGO_SCHOOL_DATA ; 
```

```python
%%sql select School_ID, NAME_OF_SCHOOL, SAFETY_SCORE
from CHICAGO_SCHOOL_DATA
ORDER BY 3 DESC LIMIT 5; 
```

```python
%%sql select School_id, NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
from CHICAGO_SCHOOL_DATA
ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC
limit 10;
```

```python
%%sql select School_id, NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
from CHICAGO_SCHOOL_DATA
ORDER BY AVERAGE_STUDENT_ATTENDANCE
limit 5 ;
```

```python
%%sql select School_id, NAME_OF_SCHOOL, REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%','') AS AVG_ATTENDENCE
from CHICAGO_SCHOOL_DATA
where AVG_ATTENDENCE < '70' ;
```

```python
%%sql SELECT COMMUNITY_AREA_NAME,  COMMUNITY_AREA_NUMBER, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT
FROM CHICAGO_SCHOOL_DATA 
group by COMMUNITY_AREA_NAME;
```

```python
%%sql SELECT COMMUNITY_AREA_NAME,  COMMUNITY_AREA_NUMBER, SUM(COLLEGE_ENROLLMENT) AS TOTAL_ENROLLMENT
FROM CHICAGO_SCHOOL_DATA 
group by COMMUNITY_AREA_NAME
ORDER BY TOTAL_ENROLLMENT;
```

```python
%%sql SELECT School_ID, NAME_OF_SCHOOL, SAFETY_SCORE
FROM CHICAGO_SCHOOL_DATA
WHERE SAFETY_SCORE<>'None'
ORDER BY SAFETY_SCORE
limit 5;
```

```python
%sql select name from sqlite_master where type = 'table' ;
```

```python
%sql SELECT name , type, length(type) from PRAGMA_TABLE_INFO ('CENSUS_DATA')
```

```python
%%sql select HARDSHIP_INDEX
from CENSUS_DATA CD, CHICAGO_SCHOOL_DATA CSD
WHERE CD.COMMUNITY_AREA_NUMBER = CSD.COMMUNITY_AREA_NUMBER AND COLLEGE_ENROLLMENT = 4368;
```

```python
%%sql SELECT HARDSHIP_INDEX, COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER
FROM CENSUS_DATA
WHERE COMMUNITY_AREA_NUMBER IN (SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_SCHOOL_DATA
                               ORDER BY COLLEGE_ENROLLMENT desc
                               limit 1) ;
```
