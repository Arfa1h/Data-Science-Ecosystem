```python
import matplotlib.pyplot as plt

# Create flowchart steps
steps = [
    "I. Conceptual Design",
    "II. Level & Blockout",
    "III. Asset Production",
    "IV. Engine Integration",
    "V. Lighting & Atmosphere",
    "VI. Interactivity & Polish"
]

# Set up figure
fig, ax = plt.subplots(figsize=(8, 10))
ax.set_xlim(0, 10)
ax.set_ylim(0, 12)
ax.axis('off')

# Draw boxes and arrows
y_positions = [11, 9, 7, 5, 3, 1]

for i, (step, y) in enumerate(zip(steps, y_positions)):
    ax.text(5, y, step, ha='center', va='center',
            bbox=dict(boxstyle="round,pad=0.4", edgecolor="black"))
    if i < len(steps) - 1:
        ax.annotate("", xy=(5, y-0.7), xytext=(5, y-1.3),
                    arrowprops=dict(arrowstyle="->", lw=1.5))

# Save the image
file_path = "/mnt/data/flowchart.png"
plt.savefig(file_path, dpi=300, bbox_inches='tight')

file_path
```
