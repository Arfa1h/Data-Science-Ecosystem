```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
# Step 1: Load a dataset 
# You can replace this with any dataset file you have (CSV, Excel, etc.) 
# Example dataset - Titanic dataset (you can replace with your own) 
df = sns.load_dataset('titanic') 
# Display first few rows 
print("Dataset Sample:") 
print(df.head()) 
# Step 2: Select one numerical column (example: 'age') 
numerical_column = 'age' 
data = df[numerical_column].dropna()  # Remove missing values 
# Compute statistics 
mean = data.mean() 
median = data.median() 
mode = data.mode()[0]  # Mode might return multiple values, select first  
std_dev = data.std() 
variance = data.var() 
data_range = data.max() - data.min()
# Display computed statistics 
print(f"\nStatistics for column '{numerical_column}':") 
print(f"Mean: {mean}") 
print(f"Median: {median}") 
print(f"Mode: {mode}") 
print(f"Standard Deviation: {std_dev}") 
print(f"Variance: {variance}") 
print(f"Range: {data_range}") 
# Step 3: Visualization - Histogram 
plt.figure(figsize=(10, 5)) 
plt.hist(data, bins=20, color='skyblue', edgecolor='black') 
plt.title(f'Histogram of {numerical_column}') 
plt.xlabel(numerical_column) 
plt.ylabel('Frequency') 
plt.grid(True) 
plt.show() 
# Step 4: Visualization - Boxplot 
plt.figure(figsize=(10, 5)) 
sns.boxplot(x=data) 
plt.title(f'Boxplot of {numerical_column}') 
plt.grid(True) 
plt.show() 
# Step 5: Outlier Detection using IQR 
Q1 = data.quantile(0.25) 
Q3 = data.quantile(0.75) 
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR 
upper_bound = Q3 + 1.5 * IQR 
outliers = data[(data < lower_bound) | (data > upper_bound)] 
print(f"\nOutliers detected in '{numerical_column}':") 
print(outliers.values) 
# Step 6: Select a categorical column (example: 'class') 
categorical_column = 'class' 
category_counts = df[categorical_column].value_counts() 
# Step 7: Visualization - Bar Chart 
plt.figure(figsize=(8, 5)) 
category_counts.plot(kind='bar', color='lightgreen', edgecolor='black') 
plt.title(f'Category Frequency - {categorical_column}') 
plt.xlabel(categorical_column) 
plt.ylabel('Frequency') 
plt.grid(True) 
plt.show() 
# Visualization - Pie Chart (optional alternative) 
plt.figure(figsize=(8, 8)) 
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140) 
plt.title(f'Category Distribution - {categorical_column}') 
plt.ylabel(' ')  # Hide y-label for better appearance 
plt.show() 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
# Load the dataset (Iris dataset from seaborn) 
df = sns.load_dataset('iris') 
# Display first few rows to understand the dataset 
print("First few rows of dataset:") 
print(df.head()) 
# Select two numerical columns for scatter plot and Pearson correlation 
x = df['sepal_length'] 
y = df['sepal_width'] 
# Scatter plot of sepal length vs sepal width 
plt.figure(figsize=(8, 5)) 
plt.scatter(x, y, color='skyblue', edgecolor='k') 
plt.title('Scatter Plot: Sepal Length vs Sepal Width') 
plt.xlabel('Sepal Length (cm)') 
plt.ylabel('Sepal Width (cm)') 
plt.grid(True) 
plt.show() 
# Calculate Pearson correlation coefficient between these two variables 
pearson_corr = x.corr(y) 
print(f"Pearson Correlation Coefficient between Sepal Length and Sepal Width: {pearson_corr:.2f}") 
# Select only numeric columns 
df_numeric = df.select_dtypes(include=['number']) 
# Compute covariance matrix 
cov_matrix = df_numeric.cov() 
# Print the result 
print("\nCovariance Matrix:") 
print(cov_matrix) 
# Compute correlation matrix 
corr_matrix = df.corr(numeric_only=True) 
print("\nCorrelation Matrix:") 
print(corr_matrix) 
# Visualize correlation matrix using heatmap 
plt.figure(figsize=(10, 6)) 
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Correlation Matrix Heatmap') 
plt.show() 
```
