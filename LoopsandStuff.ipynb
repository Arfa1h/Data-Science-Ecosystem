```python
squares = ["Blue","Green","Black","Purple"];
r = len(squares)
for i in range(r):
    print(f"color of square before is {squares[i]}")
    squares[i]="white"
    print(f"color of square after is {squares[i]}")
    print("<------------>")
    i=i+1
for i, dabba in enumerate(squares):
    print("Dabba is",i,dabba)
```
