```python
group = df[["GPU","CPU_core","Price"]]
grouptst = group.groupby(["GPU","CPU_core"], as_index = False).mean()
grouptst
```

```python
grouppivot = grouptst.pivot(index = "GPU", columns = "CPU_core")
grouppivot
```

```python
plt.pyplot.pcolor(grouppivot, cmap = 'RdBu')
plt.pyplot.colorbar()
plt.pyplot.show()
```

```python
fig, ax = plt.pyplot.subplots()
im = ax.pcolor(grouppivot, cmap='RdBu')

#label names
row_labels = grouppivot.columns.levels[1]
col_labels = grouppivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouppivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouppivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.pyplot.xticks(rotation=90)

fig.colorbar(im)
plt.pyplot.show()
```

```python
list = ['RAM_GB','CPU_frequency','Storage_GB_SSD','Screen_Size_inch','Weight_pounds','CPU_core','OS','GPU','Category']
for value in list:
    coeff, pval = stats.pearsonr(df[value], df["Price"])
    print(f"\nthe pearson coefficient between {value} and Price is {coeff} and p-value is {pval}\n")
```
