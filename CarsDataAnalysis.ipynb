```python
df_group_one = df[["drive-wheels","body-style","price"]]
```

```python
df_grouped = df_group_one.groupby(['drive-wheels','body-style'], as_index = False).agg({'price':'mean'})
df_grouped
```

```python
df_grouptst = df_group_one.groupby(['drive-wheels','body-style'], as_index = False).mean()
df_grouptst
```

```python
group_pivot = df_grouptst.pivot(index='drive-wheels',columns = 'body-style')
group_pivot
```

```python
group_pivot = group_pivot.fillna(0)
group_pivot
```

```python
new_group = df[["body-style","price"]]
grp = new_group.groupby(["body-style"],as_index=False).mean()
```

```python
grp
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pcolor(group_pivot, cmap = 'RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(group_pivot, cmap='RdBu')

#label names
row_labels = group_pivot.columns.levels[1]
col_labels = group_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(group_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(group_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
